```python
lats_longs_pd = doc[['LATITUDE','LONGITUDE']]
lats_longs = []
for index, row in lats_longs_pd.iterrows():
    latitude,longitude = row
    lats_longs.append((latitude,longitude))
#     print(latitude,longitude)
unique_lats_longs = list(set(lats_longs))
# list(zip(lats_longs['LATITUDE'], lats_longs['LONGITUDE']))
```

```python
from geopy.geocoders import Nominatim
import time

def get_local_locations():
    location_arr = []
    geolocator = Nominatim(user_agent="specify_your_app_name_here")
    i = 0 
    for lat_long in unique_lats_longs:
        i += 1
        try: 
            location = str(geolocator.reverse(str(lat_long[0])+","+str(lat_long[1])))

            if 'NYC' in location or 'New York City' in location or 'Manhattan' in location:
                location_arr.append("lat_long: " + str(lat_long) + ", location: "+str(location))
        except:
            print("error at: itr: "+i+" lat_long: "+lat_long)
    return location_arr
```
